## I'm going to remove 1603 and Doe Loan guarantee program projects and see what happens.  I think there will be a more pronounced spike in what remains

```python
import pandas as pd
```

```python
# list of operating projects in 2018
gen = pd.read_csv('/Users/rachelanderson/Dropbox (Princeton)/Data for Tax Equity Project/eia_data/eia8602018/merged_new.csv')
gen = gen[~gen['utility_name'].isna()]
gen['nameplate_cap'] = pd.to_numeric(gen['nameplate_capacity_mw'].astype(str).str.replace(',',''))
```

```python
#list of funded capacity 
grant_recips = pd.read_excel('../../../Data for Tax Equity Project/treasury_data/grant_awards.xlsx',header=1)
```

```python
#create variables for receiving 1603 loan grant or 1705 

#DoE 1705 utility-scale recipients
searchfor = ['Agua Caliente', 'California Valley Solar', 
             'Desert Sun','Mesquite Solar 1', 'AV Solar',
            'Cogentrix', 'Crescent Dune','Ivanpah',
            'Genesis Solar Energy Project',
            'Mojave Solar Project',
            'Solana', ]

gen['1603_grant'] = gen['plant_name'].isin(grant_recips['Business Name']) | gen['utility_name'].isin(grant_recips['Business Name'])
gen['1705_grant'] = gen['plant_name'].str.contains('|'.join(searchfor))
```

```python
gen[gen['plant_name'].str.contains('Solana')][['utility_name','plant_state','plant_name','nameplate_cap','technology']]
```

```python
arra_names = gen[gen['plant_name'].isin(grant_recips['Business Name']) | gen['utility_name'].isin(grant_recips['Business Name'])]['plant_name'].values
doe_names = gen[gen['plant_name'].str.contains('|'.join(searchfor))]['plant_name'].values
loan_dict = dict.fromkeys(arra_names, '1603 grant')
loan_dict.update(dict.fromkeys(doe_names, '1705 program'))
```

```python
gen['grant_program'] = gen['plant_name'].map(loan_dict)
gen['grant_program'] = gen['grant_program'].fillna("none")
```
