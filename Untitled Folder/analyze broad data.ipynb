```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from colorspacious import cspace_converter
from collections import OrderedDict
# %matplotlib inline
plt.rcParams['figure.figsize'] = [25, 20]
plt.rcParams['legend.fontsize'] = 'x-large'
plt.style.use('ggplot')
cmaps = OrderedDict()
cmaps['Qualitative'] = ['Pastel1', 'Pastel2', 'Paired', 'Accent',
                        'Dark2', 'Set1', 'Set2', 'Set3',
                        'tab10', 'tab20', 'tab20b', 'tab20c']
```

```python
raw_gen = pd.read_excel('../../Data for Tax Equity Project/eia_summary_data/total_renew_gen.xlsx', 
                                header = 3)
raw_gen.columns = [x.replace("\n", " ") for x in raw_gen.columns]
raw_gen.to_csv('../../Data for Tax Equity Project/eia_summary_data/total_renew_gen.csv')
```

```python
ann_total_renew_gen = raw_gen.loc[1:11,:]

tech_to_ignore = ['Estimated Solar Photovoltaic','Estimated Total Solar Photovoltaic','Estimated Total Solar','Total Renewable Generation at Utility Scale Facilities']
```

```python
total_gen_long = pd.melt(ann_total_renew_gen, id_vars=['Period'], var_name = 'technology', value_name='net_gen')
total_gen_long = total_gen_long[~total_gen_long['technology'].isin(tech_to_ignore)]

total_gen_long.index = pd.to_datetime(total_gen_long['Period'], format='%Y')
total_gen_long.drop(columns='Period',inplace=True)
total_gen_long.groupby(['technology','Period']).net_gen.sum().unstack(level=0).plot(kind='area',
                                            stacked=True, 
                                            grid=False,
                                            colormap=cm.get_cmap('Set2'),
                                            fontsize=18)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Annual Generation (MW)', fontsize=20)
plt.title('Total Renewable Generation at Utility Scale Facilities',fontdict = {'fontsize' : 30})
plt.savefig('total_renew_gen.png')
```
